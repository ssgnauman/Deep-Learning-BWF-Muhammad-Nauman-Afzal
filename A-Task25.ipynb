```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
# Generate some data
np.random.seed(0)
x = np.linspace(0, 1, 100)
y = np.sin(2 * np.pi * x) + np.random.normal(0, 0.2, len(x))
```

```python
# Plot the data
plt.scatter(x, y)
```

```python
# Underfitting example
model = LinearRegression()
model.fit(x[:, np.newaxis], y)
y_pred = model.predict(x[:, np.newaxis])
plt.plot(x, y_pred, label='Underfitting')
```

```python
# Overfitting example
model = make_pipeline(PolynomialFeatures(degree=15), LinearRegression())
model.fit(x[:, np.newaxis], y)
y_pred = model.predict(x[:, np.newaxis])
plt.plot(x, y_pred, label='Overfitting')
```

```python
# Regularization example
model = make_pipeline(PolynomialFeatures(degree=15), Ridge(alpha=0.1))
model.fit(x[:, np.newaxis], y)
y_pred = model.predict(x[:, np.newaxis])
plt.plot(x, y_pred, label='Regularization')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generate some data
np.random.seed(0)
x = np.linspace(0, 1, 100)
y = np.sin(2 * np.pi * x) + np.random.normal(0, 0.2, len(x))

# Plot the data
plt.scatter(x, y)

# Underfitting example
model = LinearRegression()
model.fit(x[:, np.newaxis], y)
y_pred = model.predict(x[:, np.newaxis])
plt.plot(x, y_pred, label='Underfitting')

# Overfitting example
model = make_pipeline(PolynomialFeatures(degree=15), LinearRegression())
model.fit(x[:, np.newaxis], y)
y_pred = model.predict(x[:, np.newaxis])
plt.plot(x, y_pred, label='Overfitting')

# Regularization example
model = make_pipeline(PolynomialFeatures(degree=15), Ridge(alpha=0.1))
model.fit(x[:, np.newaxis], y)
y_pred = model.predict(x[:, np.newaxis])
plt.plot(x, y_pred, label='Regularization')

plt.legend()
plt.show()
```
