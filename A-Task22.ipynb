# Example 

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

```

```python
# Load the iris dataset
iris = datasets.load_iris()

```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)

```

```python
# Create a k-NN classifier
knn = KNeighborsClassifier(n_neighbors=3)

```

```python
# Train the classifier using the training data
knn.fit(X_train, y_train)

```

```python
# Test the classifier on the test data and print the accuracy
print(knn.score(X_test, y_test))
```

<small>This code loads the iris dataset, splits it into training and test sets, creates a k-NN classifier, trains it on the training data, and then tests it on the test data to see how accurate it is.</small>

# Sample Classification code Example 

<small>This code loads the iris dataset, splits it into training and test sets, creates a decision tree classifier, trains it on the training data, and then tests it on the test data to see how accurate it is</small>

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python

# Load the iris dataset
iris = datasets.load_iris()
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
# Create a decision tree classifier
clf = DecisionTreeClassifier()
```

```python
# Train the classifier using the training data
clf.fit(X_train, y_train)
```

```python
# Test the classifier on the test data and print the accuracy
print(clf.score(X_test, y_test))
```
