```python
import pandas as pd

# Create an empty DataFrame to store the data
data = pd.DataFrame()

# Loop over the file numbers
for i in range(10):
    # Generate the file name
    file = 'states{}.csv'.format(i)
    
    # Read the file into a DataFrame
    df = pd.read_csv(file)
    
    # Append the data to the main DataFrame
    data = data.append(df)

# Display the first five rows of data
print(data.head())
```

# Understanding Training, Validation and Test Sets

<small>When teaching a computer using machine learning, you need to give it data to learn from. This data is called a dataset. To make sure the computer is learning well, you split the dataset into three parts: a training set, a validation set, and a test set.

The training set is used to teach the computer. The computer looks at this data and tries to learn from it.

The validation set is used to check how well the computer is learning. You show the computer this data and see if it can make good predictions. If the computer is not doing well, you can make changes to help it learn better.

The test set is used to see how well the computer can use what it learned on new data. You show the computer this data after it has finished learning. This helps you see if the computer can make good predictions on data it has never seen before.

Splitting the data like this helps you make sure that the computer is really learning and not just memorizing the answers</small>

# Overfitting 

<small>Overfitting is when a machine learning model is too complex and memorizes the training data instead of learning to generalize from it. This means that the model will do very well on the training data, but it will not do well on new data that it has never seen before.

Think of it like studying for a test. If you memorize all the answers to the practice questions, you will do very well on the practice test. But if the real test has different questions, you will not do well because you didnâ€™t learn how to solve the problems, you just memorized the answers.

To avoid overfitting, you can use techniques like simplifying the model or using a validation set to check how well the model is learning.</small>

# example of how you can split a dataset into training, validation, and test sets using the train_test_split function from the scikit-learn library in Python:

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
# Load sample data
iris = load_iris()
X, y = iris.data, iris.target
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Split training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Create a random forest classifier
clf = RandomForestClassifier()

# Train the model using the training set
clf.fit(X_train, y_train)

# Make predictions on the validation set
y_pred = clf.predict(X_val)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_val, y_pred)

# Print the accuracy
print(f'Validation accuracy: {accuracy:.2f}')
```
